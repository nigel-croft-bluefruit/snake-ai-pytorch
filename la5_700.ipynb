```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
snake = pd.read_csv('la5_700.csv')
snake['Efficiency'] = snake['Score'] / snake['Steps']
snake.head()
```

```python
def moving_average(x, w):
    return np.convolve(x, np.ones(w), 'valid') / w
```

```python
plt.style.use('seaborn-whitegrid')
plt.plot(snake['Score'])
plt.plot(moving_average(snake['Score'], 20))
plt.title('Score')
```

```python
plt.plot(snake['Steps'])
plt.plot(moving_average(snake['Steps'], 20))
plt.title('Steps')
```

```python
plt.plot(snake['Efficiency'])
plt.plot(moving_average(snake['Efficiency'], 20))
plt.title('Efficiency')
```
